```python
import math as mat
import numpy as np
import matplotlib.pyplot as mpl
from basic_units import radians, degrees

# MatPlotLib set fonts
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = ['DejaVu Serif']

# MatPlotLib set LaTeX use
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{siunitx}'
```

```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return '0'
    elif N == 1:
        return r'$\pi/2$'
    elif N == 2:
        return r'$\pi$'
    elif N % 2 > 0:
        return r'${0}\pi/2$'.format(N)
    else:
        return r'${0}\pi$'.format(N // 2)
```

```python
x = np.linspace(0,2*mat.pi,100)

fig, ax = mpl.subplots()
fig.suptitle(r'Plot test')

ax.plot(x,np.sin(x), xunits=radians, label=r'$\sin(x)$')
ax.plot(x,np.cos(x), xunits=radians, label=r'$\cos(x)$')
ax.xaxis.set_major_locator(mpl.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(mpl.MultipleLocator(np.pi / 4))
ax.xaxis.set_major_formatter(mpl.FuncFormatter(format_func))
ax.set_xlabel('$x$')
ax.legend()
ax.grid()
fig.savefig('plot.pdf')
```
